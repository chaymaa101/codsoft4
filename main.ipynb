```python
normal_sample = legit.sample(n=492)            #on sÃ©lectionne un Ã©chantillon alÃ©atoire de 492 lignes pour equilibrage de dataset
```

```python
# Fusionner les datasets
new_dataset = pd.concat([normal_sample, fraud], axis=0)

# Afficher le nouveau dataset
print(new_dataset['Class'].value_counts())
```

```python
new_dataset.groupby('Class').describe()
```

```python
# Ici, nous pouvons supprimer la colonne 'Time' et utiliser Ã  la place une colonne dÃ©rivÃ©e
# en utilisant la fonction timedelta de pandas pour reprÃ©senter la durÃ©e=la diffÃ©rence entre deux valeurs de temps)
delta_time = pd.to_timedelta(new_dataset['Time'], unit='s')

# CrÃ©er la colonne dÃ©rivÃ©e 'time_hour' reprÃ©sentant les heures extraites des objets timedelta
new_dataset['time_hour'] = (delta_time.dt.components.hours).astype(int)

# Maintenant, supprimer la colonne 'Time' car elle n'est plus nÃ©cessaire
new_dataset.drop(columns='Time', axis=1, inplace=True)
```

```python
new_dataset
```

```python
x = new_dataset.drop('Class', axis=1) # Supprimer la colonne 'Class' pour le dataframe x
y = new_dataset['Class']  #definir la variable cible =class 
x.shape ,y.shape
```

```python
from sklearn.model_selection import train_test_split #on divise data into test and train one
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 3, stratify = y)
```

```python
cols = list(x.columns.values)
print(cols)
```

```python
#accumulating all the column names under one variable
cols = list(x.columns.values)
```

```python
normal_entries = new_dataset.Class==0
fraud_entries = new_dataset.Class==1

plt.figure(figsize=(20,70))
for n, col in enumerate(cols):
    plt.subplot(10,3,n+1)
    sns.histplot(x[col][normal_entries], color='green', kde = True, stat = 'density')
    sns.histplot(x[col][fraud_entries], color='red', kde = True, stat = 'density')
    plt.title(col, fontsize=17)
plt.show()
```

```python
# Standardisation des donnÃ©es
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
# EntraÃ®ner le modÃ¨le de rÃ©gression logistique
model = LogisticRegression()
model.fit(x_train_scaled, y_train)

# PrÃ©dictions sur l'ensemble de test
y_pred = model.predict(x_test_scaled)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
# Ã‰valuer le modÃ¨le
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

```

```python
# Affichage de la matrice de confusion sous forme de heatmap
plt.figure(figsize=(5, 5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Matrice de Confusion')
plt.xlabel('actual val')
plt.ylabel('predited val')
plt.show()
```

```python
print("\nClassification Report:")
class_report= classification_report(y_test, y_pred)
print(class_report)
```

PrÃ©cision pour la classe 0 : 91% des prÃ©dictions pour la classe 0 sont correctes.
Rappel pour la classe 0 : 97% des exemples rÃ©els de la classe 0 sont correctement prÃ©dits.

PrÃ©cision pour la classe 1 : 97% des prÃ©dictions pour la classe 1 sont correctes.

Moyenne Macro : Moyenne non pondÃ©rÃ©e des mÃ©triques pour chaque classe (prÃ©cision, rappel, F1-score).
Moyenne PondÃ©rÃ©e : Moyenne pondÃ©rÃ©e des mÃ©triques pour chaque classe, pondÃ©rÃ©e par le support.

```python

print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, auc

# PrÃ©dictions de probabilitÃ© pour l'ensemble de test
y_prob = model.predict_proba(x_test_scaled)[:, 1]

# Calcul de la courbe ROC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# Calcul de l'AUC
roc_auc = auc(fpr, tpr)

# Affichage de la courbe ROC
plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```

la courbe ROC est proche du coin supÃ©rieur gauche, indiquant une bonne performance.
la courbe AUC est de 0.98, ce qui signifie que le modÃ¨le a une bonne capacitÃ© Ã  distinguer entre les classes positives et nÃ©gatives.
